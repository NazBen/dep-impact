```python
import numpy as np
import openturns as ot
from dependence import ImpactOfDependence
 
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
dim = 3
alpha = 0.05 
threshold = 2.
N = 20
n = 1000
measure = "KendallTau"
margins = [ot.Weibull(), ot.Normal(), ot.Normal()]

families = np.zeros((dim, dim), dtype=int)
families[1, 0] = 1
families[2, 0] = 0
families[2, 1] = 3
```

```python
def add_function(x):
    """
    """
    return x.sum(axis=1)
```

```python
impact = ImpactOfDependence(model_func=add_function, margins=margins, families=families)

impact.run(n_dep_param=N, n_input_sample=n, fixed_grid=True, 
           dep_measure=measure, seed=0)
```

```python
proba_result = impact.compute_probability(threshold)
```

```python
proba_result.draw()
```

```python
id_max = proba_result.quantity.argmax()
print id_max, proba_result.quantity[id_max]
```

```python
impact.draw_matrix_plot(id_max, figsize=(10, 10))
```

```python
impact.save_data()
```
