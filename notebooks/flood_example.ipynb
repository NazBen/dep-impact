# Impact of dependencies : Application on overflow application

This document is inspired by the application of [B. Iooss & P. Lemaître](https://hal.archives-ouvertes.fr/hal-00975701/document). 

The flood example is a reliability problem based on the equation 

$$
    S = Z_v + H + H_d - C_b \quad \mbox{with}\quad H = \left( \frac{Q}{ B K_s \sqrt{\frac{Z_m - Z_v}{L}}} \right)^{0.6} .
$$

See the article [B. Iooss & P. Lemaître](https://hal.archives-ouvertes.fr/hal-00975701/document) for the information on the input variables and their probability distributions.

We aim to study the probability of an anual overflow (when $S > 0$). In a first time, the input variables are all considered independent. In a second time, we study the impact of dependencies between variables on the overflow probabilities.

## The overflow model

We first declare the python function to compute the overflow $S$.

```python
import numpy as np
%load_ext autoreload

def func_overflow(X, model=1):
    """
    X : input variables, shape : N x 8
    """
    X = np.asarray(X)
    if X.shape[0] == X.size: # It's a vector
        n = 1
        dim = X.size
        ids = None
    else:
        n, dim = X.shape
        ids = range(n)
    assert dim == 8, "Incorect dimension : dim = %d != 8" % dim
    
    Q = X[ids, 0]
    Ks = X[ids, 1]
    Zv = X[ids, 2]
    Zm = X[ids, 3]
    Hd = X[ids, 4]
    Cb = X[ids, 5]
    L = X[ids, 6]
    B = X[ids, 7]
    
    H = (Q / (B * Ks * np.sqrt((Zm - Zv) / L)))**1.
    S = Zv + H - Hd - Cb
    
    if model == 1:
        return S
    elif model == 2:
        Cp = (S > 0.) + (0.2 + 0.8 * (1. - np.exp(-1000. / (S**4)))) * (S <= 0.) + 1./20. * (Hd * (Hd > 8.) + 8*(Hd <= 8.))
        return Cp
```

We then atribute distributions to the input variables.

```python
import openturns as ot
tmp = ot.Gumbel(1013., 558., ot.Gumbel.MUSIGMA)
# tmp = ot.Gumbel(1013., 558., ot.Gumbel.ALPHABETA)
dist_Q = ot.TruncatedDistribution(tmp, 500., 3000.)
dist_Ks = ot.TruncatedNormal(30., 8., 15., np.inf)
dist_Zv = ot.Triangular(49., 50., 51.)
dist_Zm = ot.Triangular(54., 55., 56.)
dist_Hd = ot.Uniform(7., 9.)
dist_Cb = ot.Triangular(55., 55.5, 56.)
dist_L = ot.Triangular(4990., 5000., 5010.)
dist_B = ot.Triangular(295., 300., 305.)

margins = [dist_Q, dist_Ks, dist_Zv, dist_Zm, dist_Hd, dist_Cb, dist_L, dist_B]
var_names = ["Q", "K_s", "Z_v", "Z_m", "H_d", "C_b", "L", "B"]
```

And we draw the distributions.

```python
from openturns.viewer import View
import matplotlib.pyplot as plt
%matplotlib inline

input_dim = 8
NUM_COL = 4
NUM_ROW = input_dim/NUM_COL

fig, axes = plt.subplots(NUM_ROW, NUM_COL, figsize=(NUM_COL*3, NUM_ROW*3))

for k, dist in enumerate(margins):    
    ax = axes.ravel()[k]
    data = np.asarray(dist.drawPDF().getDrawable(0).getData())
    x = data[:, 0]
    y = data[:, 1]
    ax.plot(x, y)
    ax.axis("tight")
    ax.set_xlabel("$%s$" % var_names[k])
    fig.tight_layout()
```

```python
families = np.zeros((input_dim, input_dim))
families[1, 0] = 1

impact_flood = ImpactOfDependence(func_overflow, margins, families)

n_param = 100
n_input_sample = 10000
fixed_grid = True
impact_flood.run(n_param, n_input_sample, fixed_grid)
```

```python
cops = [1, 3, 13, 23, 33]

n_param = 100
n_input_sample = 1000
fixed_grid = True
alpha = 0.99

quantiles = []
params = []
for cop in cops:
    families = np.zeros((input_dim, input_dim))
    families[1, 0] = cop
    
    impact_flood = ImpactOfDependence(func_overflow, margins, families)
    impact_flood.run(n_param, n_input_sample, fixed_grid)
    conv = impact_flood._copula[impact_flood._corr_vars[0]]
    quant_result = impact_flood.compute_quantiles(alpha)
    params.append(conv.to_Kendall(quant_result.cond_params))
    quantiles.append(quant_result.quantity)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
for i, (param, quantile) in enumerate(zip(params, quantiles)):
    ax.plot(param, quantile, '-', linewidth=2, label="%d" % cops[i])
    
# ax.set_xlim(-1., 0.)
ax.legend(loc='best')

fig.tight_layout()
```

```python
cops = [1, 3, 13, 23, 33]

n_param = 50
n_input_sample = 10000
fixed_grid = True
alpha = 0.95

all_quantiles = {}
all_param = {}
for i in range(input_dim):
    for j in range(i):
        quantiles = []
        params = []
        for cop in cops:
            families = np.zeros((input_dim, input_dim))
            families[i, j] = cop

            impact_flood = ImpactOfDependence(func_overflow, margins, families)
            impact_flood.run(n_param, n_input_sample, fixed_grid)
            conv = impact_flood._copula[impact_flood._corr_vars[0]]
            quant_result = impact_flood.compute_quantiles(alpha)
            params.append(conv.to_Kendall(quant_result.cond_params))
            quantiles.append(quant_result.quantity)
        all_quantiles[i, j] = quantiles
        all_param[i, j] = params
```

```python
fig, axes = plt.subplots(input_dim, input_dim, figsize=(15, 15), sharex=True, sharey=True)
for i in range(input_dim):
    for j in range(input_dim):
        ax = axes[i, j]
        if i > j:
            quantiles = all_quantiles[i, j]
            params = all_param[i, j]
            for k, (param, quantile) in enumerate(zip(params, quantiles)):
                ax.plot(param, quantile, '-', linewidth=2, label="%d" % cops[k])
        ax.set_yticks([])
        ax.set_xticks([])
# ax.legend(loc='best')

fig.tight_layout()
```
